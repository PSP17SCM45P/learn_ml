# Classifying Iris Species

We'll take a look at the popular Iris plant multiclass classification problem (3 discrete valued labels).

```python
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from IPython.display import display
%matplotlib inline
```

```python
# Iris dataset is of sklearn.datasets.base.Bunch (very similar to dictionary)
dataset = load_iris()
print(type(dataset))
```

```python
# Keys of load_iris()
print("Keys of Iris:\n{}".format(dataset.keys()))
```

```python
# Print the description of the dataset, acess "DESCR" key
print(dataset["DESCR"][:200])
print("... NOTE: expand string selection for more ...")
```

```python
# Print the possible label names (the discrete values)
labels = dataset["target_names"]
print(labels)
print("Label attribution: {}".format(((0, labels[0]), (1, labels[1]), (2, labels[2]))))
```

```python
# Print the input feature names (column names of training set)
print(dataset["feature_names"])
```

```python
# Print the dataset's "data" -- values of the features
X_matrix = dataset["data"]
print("4 samples of Iris dataset (without labels)\n{}\n".format(X_matrix[0:4, :]))
print("Shape of Iris dataset: {}".format(X_matrix.shape))
print("Data point count: {}".format(X_matrix.shape[0]))
print("Feature count: {}".format(X_matrix.shape[1]))
```

```python
# Print the dataset's "target" -- values of the labels
Y_matrix = dataset["target"]
print("Full labels array:\n{}\n".format(Y_matrix))
print("Shape of labels vector: {}".format(Y_matrix.shape))
print("Size of labels vector: {}\n".format(Y_matrix.size))
# From previous cell
print("Label attribution: {}".format(((0, labels[0]), (1, labels[1]), (2, labels[2]))))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_matrix, Y_matrix, test_size=0.25, \
                                                   random_state=0)
```

```python
# Observe the shapes of the resulting splits
print("Shape of X_train: {}".format(X_train.shape))
print("Labels count of y_train: {}\n".format(y_train.size))

print("Shape of X_test: {}".format(X_test.shape))
print("Labels count of y_test: {}".format(y_test.size))
```

```python
# Visualizing the data --> pair plots
iris_train_dataframe = pd.DataFrame(data=X_train, columns=dataset["feature_names"])
print("Dataframe representation of X_train (required for pair plotting): \n{}\n"\
      .format(iris_train_dataframe.head()))

from pandas.plotting import scatter_matrix
scatter_matrix(iris_train_dataframe, c=y_train, figsize=(7, 7), marker="o", \
              hist_kwds={"bins": 20}, s=60, alpha=0.8, cmap=mglearn.cm3)
# alpha: color thickness, s: individual dot size, marker: shape of mark, cmap: ColorMap settings
```

```python
''' Apply a classification algorithm: K-Nearest Neighbors 
    - k: how many neighbors to compare test point against 
    (use k=1 for now) '''
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=1)
knn_classifier.fit(X_train, y_train)
```

```python
# Make a prediction with a dummy sample
dummy_sample = np.array([[5, 2.9, 1, 0.2]])
dummy_prediction = knn_classifier.predict(dummy_sample)
print("Label attribution prediction: {}".format(dummy_prediction))

translated_prediction = dataset["target_names"][dummy_prediction]
print("Actual prediction of Iris specie: {}".format(translated_prediction))
```

```python
# Evaluating Model Performance
X_test_predictions = knn_classifier.predict(X_test)
accuracy = np.mean(X_test_predictions == y_test)
print("knn classifier predictions on x_test vs y_test: {}".format(accuracy))
```

```python
# Using knn_classifier.score() --> initiates a predict for us on X_test, then compares to y
knn_classifier.score(X_test, y_test)
```
